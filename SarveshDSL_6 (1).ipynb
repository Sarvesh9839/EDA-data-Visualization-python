```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load data


marks=pd.read_csv('marks_univariate.csv')
```

```python
# Histogram using Matplotlib
import matplotlib.pyplot as plt
plt.hist(marks['Marks'], bins=10, color='skyblue', edgecolor='black')
plt.title("Histogram of Marks (Matplotlib)")
plt.xlabel("Marks")
plt.ylabel("Frequency")
plt.show()
```

```python
#Bivariate (Scatter plot)
import seaborn as sns
import pandas as pd
tips = pd.read_csv("tips_small.csv")

# Scatter plot with Seaborn
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='sex')
plt.title("Scatter Plot of Total Bill vs Tip Amount")
plt.xlabel("Total Bill")
plt.ylabel("Tip Amount")
plt.show()

# Bar Plot (Multivariate)

tit = pd.read_csv("titanic_dataset.csv")

# Simple Bar Plot
sns.barplot(data=tit, x='Embarked', y='Fare', palette='rainbow')
plt.title("Average Fare by Embarked Town")
plt.show()

# Multivariate Bar Plot with hue
sns.barplot(data=tit, x='Embarked', y='Fare', hue='Pclass', palette='rainbow')
plt.title("Fare by Embarked Town, Divided by Class")
plt.show()

# Violin Plot

sns.violinplot(data=tit, x='Pclass', y='Fare')
plt.title("Violin Plot of Fare by Pclass")
plt.show()

#Box Plot

sns.boxplot(data=tit, x='Pclass', y='Fare')
plt.title("Box Plot of Fare by Pclass")
plt.show()

```

```python
import pandas as pd
import matplotlib.pyplot as plt

line = pd.read_csv("titanic_dataset.csv")
plt.plot(line['Age'], line['Fare'], label='Age vs Fare')
plt.title("Line Plot of Age vs Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.legend()
plt.show()
```
